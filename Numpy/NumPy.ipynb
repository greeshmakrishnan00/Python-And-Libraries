# NumPy Filter Array

```python
import numpy as np #create an array from the element on index 0 and 2
arr=np.array([41,42,43,44])
x=[True,False,True,False]
newarr=arr[x]
print(newarr)
```

```python
#create a fliter  array that will return only value higher than 42:

import numpy as np
arr=np.array([41,42,43,44])
filter_arr=[]
for element in arr:
  if element >42:
    filter_arr.append(True)
  else:
    filter_arr.append(False)
newarr=arr[filter_arr]
print(filter_arr)
print(newarr)
```

## Range, variance, Standard deviation, IQR  

```python
import numpy as np
import scipy.stats as stats
data=[12,15,14,10,18,20,22,25,14,17]

#range

range_value=max(data)-min(data)
print(f"Range :{range_value}")

#variance

variance_value=np.var(data,ddof=0)
sample_variance_value=np.var(data,ddof=1)
print(f"population Variance :{variance_value}")
print(f"sample Variance : {sample_variance_value}")

#standard deviation

std_dev_value=np.std(data,ddof=0)
sample_std_dev_value=np.std(data,ddof=1)
print(f"Population Standard Deviation : {std_dev_value}")
print(f"Sample standard Devition :{sample_std_dev_value}")

#IQR
Q1=np.percentile(data,25)
Q3=np.percentile(data,75)
iqr_value=Q3-Q1
print(f"interquartile Range (IQR):{iqr_value}")
```

# Union

```python
import numpy as np
arr1=np.array([1,2,3,4])
arr2=np.array([3,4,5,6])
newarr=np.union1d(arr1,arr2)
print(newarr)
```

# intersection

```python
import numpy as np
arr1=np.array([1,2,3,4])
arr2=np.array([3,4,5,6])
newarr=np.intersect1d(arr1,arr2,assume_unique=True)
print(newarr)
```

# Finding Difference

```python
#Find the difference of the set1 from set2
import numpy as np
arr1=np.array([1,2,3,4])
arr2=np.array([3,4,5,6])
newarr=np.setdiff1d(arr1,arr2,assume_unique=True)
print(newarr)
```

# Finding Symmetric Difference

```python
#find the symmetric Difference of the se1 and set2
import numpy as np
set1=np.array([1,2,3,4])
set2=np.array([3,4,5,6])
newarr=np.setxor1d(set1,set2,assume_unique=True)
print(newarr)
```
